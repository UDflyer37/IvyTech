# ASSIGNMENT 1: Triangles

<br>A triangle has three sides where the sum of the length of any (every) two of the sides must exceed the other side.<br>

<br>In a right triangle, the square of the length of the longest side is equal to the sum of the squares of the lengths of the other two sides.<br>

<br>In an isosceles triangle the length of any two of the sides must be equal.<br>

<br>In an equilateral triangle the length of all three sides must be equal.<br>

<br>***_Write a program that prompts the user to enter the lengths of three sides of a triangle. The program will output a message indicating:_***<br>
- whether or not the lengths form a triangle.<br> 
- whether or not the triangle is a right triangle, an isosceles triangle, and/or an equilateral triangle.<br>
- Input Validation: Do not accept a number <= 0 for any side.<br>

```python
################################
# Program name: triangles.py
#  Author: Alexander Otto
#  Course: Python Essentials
#  Date: 31AUG2022
#  Assignment: Module 3 - Triangles
#  Purpose: Determine if the user dimensions for a triangle and what type of triangle it is.
########################################

def triangles():
    print("\nLet's make a triangle!\n")

    a = float(input("What is the length of side A? "))
    while a <= 0: a = float(input("Please use a positive integer for the length of side A! "))

    b = float(input("What is the length of side B? "))
    while b <= 0: b = float(input("Please use a positive integer for the length of side B! "))

    c = float(input("What is the length of side C? "))
    while c <= 0: c = float(input("Please use a positive integer for the length of side C! "))

    print("\nTesting to see if this is a valid triangle...")
    if a + b < c or b + c < a or a + c < b:
        return print("This is not a valid triangle! Try again!")

    print("This is a valid triangle!\n")
    
    sides = sorted([a,b,c])
    
    if sides[2]**2 == sides[0]**2 + sides[1]**2: print("This is a right triangle!")
        
    if a == b == c: print("This is an equilateral triangle!")
        
    if a == b and a != c or b == c and b != a or a == c and a != b: print("This is an isosceles triangle!")

print("TEST #1: Data Validation and Right Triangle Test")
triangles()

print("\n________________________________")
print("\nTEST #2: Equilateral Triangle Test")
triangles()

print("\n________________________________")
print("\nTEST #3: Isosceles Triangle Test")
triangles()

print("\n________________________________")
print("\nTEST #4: Improper Triangle Test")
triangles()
```

# ASSIGNMENT 2:  Distance Traveled

<br>The distance a vehicle travels can be calculated as follows:<br>

<br>distance = speed * time<br>

<br>For example, if a train travels 40 miles per hour for 3 hours, the distance traveled is 120 miles.<br>

<br>***_Write a program that asks the user for the speed of a vehicle (in miles per hour) and how many hours it has traveled. The program should then use a loop to display the distance the vehicle has traveled for each hour of that time period. Here is an example of the output:_***<br>
- What is the speed of the vehicle in mph? 40<br>
- How many hours has it traveled? 3<br>

<span style="font-size:1.25em;"> ***Hour*** | <span style="font-size:1.25em;"> ***Distance Traveled***
:-----:|:-------:
 1     |      40 
 2     |      80 
 3     |     120 

- Input Validation: Do not accept a negative number for speed and do not accept any value less than 1 for time traveled.<br>

```python
################################
# Program name: distance.py
#  Author: Alexander Otto
#  Course: Python Essentials
#  Date: 31AUG2022
#  Assignment: Module 3 - Distance Traveled
#  Purpose: Calculate Distance Traveled.....
########################################

def distance():
    print("\nLet's see how far you traveled!\n")

    speed = float(input("What is the speed of the vehicle in MPH? "))
    while speed < 0: speed = float(input("Please use a positive integer for the speed! "))

    time = int(float(input("How many hours have you traveled? ")))
    while time < 1: time = int(input("The number of hours traveled should be at least one! "))

    print("\nLet's see the hourly breakdown of distance traveled!")
    for i in range(time):
        print(f"\nAt hour {i+1} traveling at {speed} mph, you have traveled a total distance of {(i+1)*speed} miles!")

print("TEST #1: Example Numbers Test")
distance()

print("\n________________________________")
print("\nTEST #2: Testing Using Float Values")
distance()

print("\n________________________________")
print("\nTEST #3: Data Validation Test")
distance()
```
