# M01 Programming Assignment
### Phone List

___Write a program that provides a menu-driven digital contact list to the user. Create a dictionary with four or five contacts (names and numbers). The user should be able to add/update/delete entries to the contact list.___

The user should then be able to:
* retrieve the phone number for a given contact name
* add new contacts
* delete contacts
* update the phone number for a contact.

```python
import sys

contact = {"Alexander":"937-671-9346","John":"123-456-7890", "Smith":"012-345-6789", "Kyle":"123-123-1234"}

#add to contacts
def add():
    global contact
    validated = False
    
    name = str(input("Please enter the name of the new contact! ").strip())
    print(f"\nPlease confirm the name you entered is correct: {name}\nIs this correct (y / n)?")
    choice = input().strip().lower()
    if choice.startswith("n"):
        add()
        sys.exit()

    if name in contact:
        print(f"\nThis Contact Already Exists! \nHere is their infomation:\n{name}: {contact[name]}")
        
        print("Would you like to Modify this contact information? (y/n)")
        choice = input().strip().lower()
        if choice.startswith("y"):
            update()
    
    else:
        while validated == False:
            number = str(input("\nPlease enter the phone number of the new contact (eg. 1234567890)! ").strip())
            try:
                contact[name] = format(int(number[:-1]), ",").replace(",", "-") + number[-1]  #properly format phone number
                print("\nWe have saved your new contact!")
                
                print(f"\nPlease confirm the number you entered for {name} is correct: {contact[name]}\nIs this correct (y / n)?")
                choice = input().strip().lower()
                if choice.startswith("y"):
                    validated = True
                    print(f"\nThank you for confirming the contact information for {name}!")
                    
                else: 
                    print("\nLet try that again! Make sure you enter the number with no spaces or special characters!")
                
            except:
                print("\nThat is not a valid number, please try again! Enter the number with no spaces or special characters!")
    
        
    print("\nWould you like to perform any other actions? (y / n)")
    choice = input().strip().lower()
    if choice.startswith("y"):
        main() 

#Update contacts        
def update():
    global contact
    validated = False
    
    print("\nWhich contact would you like to update?")
    for name in contact:
        print(f"\t{name}: {contact[name]}")
  
    name = input().strip()
    
    if name in contact:
        while validated == False:
            number = str(input(f"\nPlease enter the new phone number for {name} (eg. 1234567890)! ").strip())
            try:
                contact[name] = format(int(number[:-1]), ",").replace(",", "-") + number[-1]  #properly format phone number
                print("\nThis contact has been updated!")
                
                print(f"\nPlease confirm the number you entered for {name} is correct: {contact[name]}\nIs this correct (y / n)?")
                choice = input().strip().lower()
                if choice.startswith("y"):
                    validated = True
                    print(f"\nThank you for confirming your update for {name}!")
                else: 
                    print("\nLet try that again! Make sure you enter the number with no spaces or special characters!")
                
            except:
                print("\nThat is not a valid number, please try again! Enter the number with no spaces or special characters!")

    else:
        print("\nSorry, that contact does not exist!")
  
    print("\nWould you like to perform any other actions? (y / n)")
    choice = input().strip().lower()
    if choice.startswith("y"):
        main()    
    
#Delete contact
def delete():
    global contact
    print("\nWhich contact would you like to delete?")
    for name in contact:
        print(f"\t{name}: {contact[name]}")
  
    name = input().strip()
  
    if name in contact:
        print(f"\nAre you sure you want to delete {name}'s contact information?(y / n)?")
        choice = input().strip().lower()
        if choice.startswith("y"):
            del(contact[name])
            print(f"\n{name}'s contact information has been deleted!")
    else:
        print("\nContact not found!")
  
    print("\nWould you like to perform any other actions? (y / n)")
    choice = input().strip().lower()
    if choice.startswith("y"):
        main()

#Find contacts
def find():
    global contact
    print("\nWho would you like to search for?")
  
    name = input().strip()
  
    if name in contact:
        print(f"\nContact Found! Here is their information:\n{name}:\t{contact[name]}")
    else:
        print("\nSorry, we do not have that contact stored!")
  
    print("\nWould you like to perform any other actions? (y / n)")
    choice = input().strip().lower()
    if choice.startswith("y"):
        main()
        
def main():
    print("\nPlease choose from one of the following options:\n\n1. Add Contact number\n2. Update Contact number\n3. Delete Contact number\n4. Find Contact number\n")
  
    try:
        choice = int(input())
        
        if 0<choice<5:
        
            choice_menu = {
                1: add,
                2: update,
                3: delete,
                4: find
            }

            choice_menu[choice]()
        else:
            print("\nThis is not a valid input . . Please pick a number from the list below!")
            raise ValueError
    except:
        main()
  
  
if __name__ == "__main__":
    print("Welcome to our menu-driven digital contact list!")
    main()
```
